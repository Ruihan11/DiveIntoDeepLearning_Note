### Save memory
- Y's address is changed (not good)

```python
before = id(Y)
Y = Y + X
id(Y) == before
```

1. use X[:] = X + Y
2. use X += Y

```python
Z = torch.zeros_like(Y)
print('id(Z):', id(Z))
Z[:] = X + Y
print('id(Z):', id(Z))
```

```python
before = id(X)
X += Y
id(X) == before
```
